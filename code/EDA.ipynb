```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from nltk.corpus import stopwords
%matplotlib inline
%config InlineBackend.figure_format ='retina'
```

# Looking at Labeled Data Only. 504 Hurricane Tweets. 502 Fire Tweets

```python
df = pd.read_csv('../Data/df8_SUPER.csv')
```

```python
df.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
df.shape
```

```python
df['disaster'].value_counts()
```

## Separating DataFrame into hurricane and fire subsets to perform EDA and draw insights

```python
hurricane = df[df['disaster']=='hurricane']
```

```python
fire = df[df['disaster']=='fire']
```

## Hurricane EDA

### TFIDF Vectorizer

```python
h_tvec = TfidfVectorizer(stop_words=stopwords.words('english'), min_df=5, max_df=1.0,ngram_range=(1,2))
```

```python
h_term_mat = h_tvec.fit_transform(hurricane['text'])
```

```python
h_term_df = pd.DataFrame(h_term_mat.toarray(), columns=h_tvec.get_feature_names())
```

Computing max TFIDF weights for all features

```python
h_wts_allwords = []

for column in h_term_df.columns:
    h_wts_allwords.append(h_term_df[column].max())
```

```python
h_tfidf_allwords = pd.DataFrame({'words':h_term_df.columns, 'weights':h_wts_allwords})
```

```python
h_tfidf_allwords.sort_values('weights', ascending=False,inplace=True)
```

Looking for vectorized words that give insights to hurricane related needs

```python
for word in h_tfidf_allwords['words']:
    print(f"'{word}',")
```

selected features

```python
features = ['free','church','children','pray','family','government','shelter','loved ones', 'relief fund',
            'services']
```

```python
hurricane_chart = h_term_df[features]
```

```python
h_wts = []

for column in hurricane_chart.columns:
    h_wts.append(hurricane_chart[column].max())
```

```python
h_tfidf = pd.DataFrame({'words':hurricane_chart.columns, 'weights':h_wts})
```

```python
h_tfidf.sort_values('weights', inplace=True)
```

```python
plt.figure(figsize=(20,15))
plt.barh(h_tfidf['words'],
        h_tfidf['weights'],
        color='#1399F6')
plt.title('TFIDF Weights for Hurricane Needs', fontsize=20)
plt.xlabel('TFIDF Weights', fontsize=20)
plt.ylabel('Hurricane Needs', fontsize=20)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
```

## Plotting Feature Correlations with Label

```python
f_term_df.insert(0, 'requesting_help',fire['requesting_help'].values)
```

```python
f_heatmap = f_term_df.corr()[['requesting_help']]
```

```python
f_corr = f_heatmap[(f_heatmap['requesting_help']>.3) | (f_heatmap['requesting_help']<-.3)]
```

```python
f_corr.sort_values('requesting_help',ascending=False,inplace=True)
```

```python
f_corr.drop('requesting_help',axis=0,inplace=True)
```

```python
# df_corr = f_term_df.corr()
plt.figure(figsize=(15,15))
sns.heatmap(f_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
# Source: https://github.com/mwaskom/seaborn/issues/1773
b, t = plt.ylim() # discover the values for bottom and top
b += 0.5 # Add 0.5 to the bottom
t -= 0.5 # Subtract 0.5 from the top
plt.ylim(b, t); # update the ylim(bottom, top) values
```
